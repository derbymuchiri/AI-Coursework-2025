```python
#input statement
days = int(input("Enter number days of your choice?"))
#formula for converting number of days into seconds
seconds = days*24*60*60
#output statement
print("The number of seconds in the days given is", seconds)
```

```python
import math
#writing the code to request the user to input the radius
radius = int(input("Enter the radius of the sphere?"))
#Defining the formula to calculate the radius
volume = (4/3)*math.pi*(radius**3)
#The print statement output
print("The volume of the sphere is",volume)
```

```python
#Program to compute the area and perimeter of a square
#Input statement
length = int(input("Enter the length of the square?"))

#Defining the function for Area of a square
def Calculate_area_square(side):
  return side * side

#defing the function for Perimeter of the square
def Calculate_perimeter_square(side):
  return side * 4

#Calling the functions
Area = Calculate_area_square(length)
Perimeter = Calculate_perimeter_square(length)

#Output Statement
print("The area of the square is",Area)
print("The perimeter of the square is",Perimeter)
```

```python
#Program to determine the character input
#input statement
character = input("Input a character?")
#Defining function to check the character input
def checking_character(char):
  if char.isupper():
    return "The character is uppercase"
  elif char.islower():
    return "The character is lowercase"
  else:
    return "This is neither uppercase nor lowercase"
#Calling the functions
Result = checking_character(character)
print(Result)
```

```python
#initializing variables x and y
x=0
y=100
#Repeating the loop
while y >= 6:
#Subtracting
  y = y - 4
#Adding (2/y) to x
x = x + (2/y)
#Printing the value of x
print("The value of x is",x)
```
