```python
import pandas as pd

# Load the uploaded CSV
df = pd.read_csv('netflix_titles.csv')

# View the first few rows
df.head()
```

```python
#question 1
south_african_movies = df[(df['type'] == 'Movie') & (df['country'].str.contains('South Africa', na=False))]
num_sa_movies = len(south_african_movies)
print(f"Number of Movies from South Africa are: {num_sa_movies}")
```

```python
#question  2
description_12th = df.loc[11, 'description']
print(f"The movie in the 12th row is:\n{description_12th}")

```

```python
#question 3

sa_movies_2021 = south_african_movies[south_african_movies['release_year'] == 2021]
print(f"South African Movies released in 2021:")
print(f"Number of movies: {len(sa_movies_2021)}")

if len(sa_movies_2021) > 0:
    print("\nMovies:")
    for idx, movie in sa_movies_2021[['title', 'release_year']].iterrows():
        print(f"   - {movie['title']} ({movie['release_year']})")
else:
    print("No South African movies found from 2021")
```
